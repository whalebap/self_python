```python
import pandas as pd
# 데이터프레임 활용을 위해 판다스 임포트.
```

```python
file = pd.read_csv("data/KNI_TC_LC_GNRL_WIK_SETLE_INFO_20191211170614.csv")
file.head()
# 9월 마지막 한 주간 모든 경기도 일반 지역화폐 사용 데이터가 담긴 csv 파일(41만 row)
# csv 파일을 읽어냄. head() 메소드로 초반 row만 확인.
```

```python
file=file[['성별코드', '연령대코드', '가맹점업종명', '결제금액']]
file.head()
# 연령대 및 성별별 가맹점 업종 사용량을 찾기 위해 필요한 칼럼만 뽑아냄.
```

```python
file=file.dropna()
file.head()
# 결제금액이 없거나 가맹점 업좀명이 없는 불필요한 row를 삭제함. 
```

```python
file.연령대코드=pd.to_numeric(file.연령대코드, errors='coerce')
# groupby를 해주기 전 확인차 성별코드 내 속성들을 숫자로 치환해서 예외사항을 막음.
```

```python
file=file.groupby(['성별코드', '연령대코드', '가맹점업종명'])
file.sum()['결제금액']
# 어떤 성별이 어떤 연령대에서 어떤 업종에 돈을 얼마나 썼는지 groupby로 확인할 수 있음. 
# sum()을 통해 총 결제금액을 알 수 있음. 
# 458row 정도로 전처리로 테이블 크기가 대폭 줄어듦.
```

```python
file.sum().reset_index().to_csv('consume.csv', index=False, header=False, encoding='utf-8-sig')
# 결과물을 csv 파일로 저장함. 성별, 연령별 소비를 쉽게 파악할 데이터 테이블을 얻게 됨. 
```
