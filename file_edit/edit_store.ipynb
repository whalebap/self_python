```python
import pandas as pd
```

```python
store=pd.read_csv("data/teamproject_store.csv")
store.head()
```

```python
industry=pd.read_csv("data/industry.csv")
industry.head()
```

```python
store=pd.merge(store, industry, how="inner", on="store_type_code")
store.head()
```

```python
store=store[['store_name', 'main_code', 'store_type_x', 'store_type_code', 'local_name', 'address', 'phone_number', 'long', 'lati', 'searchcount']]
store.head()
```

```python
store=store.rename(columns={'store_type_x':'store_type'})
store.head()
```

```python
store.to_csv("store.csv", index=False, header=False)
```
