```python
industry=pd.read_csv("data/industry.csv")
industry.head()
```

```python
store=pd.read_csv("data/store.csv")
store.head()
```

```python
store.store_type_code=pd.to_numeric(store.store_type_code, errors='coerce')
industry.store_type_code=pd.to_numeric(industry.store_type_code, errors='coerce')
```

```python
store=pd.merge(store, industry, on='store_type_code', how='inner')
store.head()
```

```python
store=store[['store_name', 'main_code', 'store_type_x', 'store_type_code', 'local_name', 'road_address', 'store_phone', 'latitude', 'longitude', 'star_ranking', 'search_result_count']]
store=store.rename(columns={'store_type_x':'store_type'})
store.head()
```

```python
store.to_csv('store(main_code).csv', index=False)
```
