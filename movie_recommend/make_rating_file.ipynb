```python
import pandas as pd
# 사용자의 가맹점 선호도로 추천을 하기 전에 데이터셋이 필요했음. 
# keggle에서 영화 메타 정보, 유저 영화 리뷰 테이블을 가져와 기반으로 삼음. 
# https://www.kaggle.com/rounakbanik/the-movies-dataset
```

```python
meta = pd.read_csv('data/movies_metadata.csv')
meta = meta.sort_values(by='id' ,ascending=True)
meta.head()
# 스토어 데이터의 기반이 되는 영화 메타 정보를 불러옴. 약 4.5만 row. 
# sort_values로 id 칼럼 순으로 정렬함. 
```

```python
meta=meta[['id']]
meta=meta.rename(columns={'id':'store_id'})
meta.head()
# 필요한 칼럼인 'id'만 남김. 
# store 테이블과 merge를 위해 같은 칼럼명으로 바꿔줌. 
```

```python
store=pd.read_csv('teamdata/store.csv')
store.head()
# 기존 가맹점 파일을 열어줌. 
```

```python
city=store['local_name']=='의정부시'
store = store[city]
store.head()
# 전체 가맹점 중 의정부시에 속한 가맹점만 선별함. 
```

```python
store=store[['id']]
store=store.rename(columns={'id':'store_id'})
store.head()
# 의정부 가맹점들도 필요한 store_id만 칼럼을 남긴 뒤 칼럼명을 바꿔줌.
```

```python
meta.store_id = pd.to_numeric(meta.store_id, errors='coerce')
store.store_id=pd.to_numeric(store.store_id, errors='coerce')
# merge 전 둘의 store_id의 데이터타입을 숫자로 맞춰줘서 에러를 막음. 

data=pd.merge(store, meta, on='store_id', how='inner')
data.head()
# 영화 테이블과 의정부 가게 테이블을 이너 조인함. 
```

```python
rating=pd.read_csv('data/ratings.csv')
rating.head()
# 유저 영화 평점 데이터를 가져옴.(2천만 row) 
```

```python
rating.describe()
# 최종 결과의 평균과 표준편차를 비교하기 위해 describe를 씀. 
```

```python
rating = rating[['userId', 'movieId', 'rating']]
rating=rating.rename(columns={'movieId':'store_id'})
rating.head()
# 필요한 칼럼만 남긴 뒤 칼럼명을 바꿔줌. 
```

```python
user=pd.read_csv("teamdata/eijungbuUser.csv")
user.head()
#의정부 유저만 있는 파일을 불러옴. 
```

```python
user=user[['userId']]
rating=pd.merge(rating, user, on='userId', how='inner')

rating=rating[['userId', 'store_id', 'rating']]
rating.sort_values(by=['userId'], axis=0, inplace=True)
rating.head()
# 의정부 유저와 rating 테이블의 user를 일치시켜 join함. 
# 테이블엔 의정부 유저로 치환된 유저들이 준 별점만 남아 있음. 
```

```python
result_data=pd.merge(rating, data, on='store_id', how='inner')

result_data=result_data[['userId', 'store_id', 'rating']]
result_data.sort_values(by=['userId'], axis=0, inplace=True)
# 의정부 가게들로 치환된 영화 테이블을 store_id를 기준으로 join함. 
# 결과적으로 의정부 주민들이 의정부 가게에 별점을 준 테이블이 생성됨. 
# 이 테이블을 userId 칼럼의 인덱스를 기준(axis=0)으로 정렬한(sort_values) 상태로 놔둠(inplace=True)
```

```python
result_data.describe()

# describe를 통해 총합, 평균(mean), 표준편차(std)를 확인함.
# 원데이터의 평균(3.55)과 표준편차(1.06)에서 크게 변화하지 않은 것을 확인함. 
```

```python
result_data.to_csv("result_rating.csv", index=False, header=False)
# 파일 저장. 
```
