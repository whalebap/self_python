```python
import numpy as np
import pandas as pd
```

```python
meta = pd.read_csv('teamdata/teamproject_store.csv')
meta.head()
```

```python
meta = meta[['id', 'store_name', 'store_type_code', 'store_type']]
meta=meta.rename(columns={'id': 'store_id'})
meta.head()
```

```python
users = pd.read_csv('teamdata/teamproject_user.csv', encoding='latin')
users = users[['id', 'store_id', 'star_ranking']]
users.head()
```

```python
users.describe()
```

```python
meta.store_id=pd.to_numeric(meta.store_id, errors='coerce')
users.store_id=pd.to_numeric(users.store_id, errors='coerce')
```

```python
data=pd.merge(users, meta, on='store_id', how='inner')
data.head()
```

```python
matrix=data.pivot_table(index='id', columns='store_name', values='star_ranking')
matrix.head(20)
```

```python
TYPE_WEIGHT=0.1
def pearsonR(s1, s2):
    s1_c=s1-s1.mean()
    s2_c=s2-s2.mean()
    return np.sum(s1_c*s2_c) / np.sqrt(np.sum(s1_c **2) * np.sum(s2_c **2))

def recommend(input_store, matrix, n, similar_type=True):
    input_types=meta[meta['store_name'] == input_store]['store_type'].iloc(0)[0]
    result=[]
    for store_name in matrix.columns:
        if store_name == input_store:
            continue
            
        cor = pearsonR(matrix[input_store], matrix[store_name])
        
        if similar_type and len(input_types)>0:
            temp_type = meta[meta['store_name']==store_name]['store_type'].iloc(0)[0]
            same_count=np.sum(np.in1d(input_types, temp_type))
            cor += (TYPE_WEIGHT * same_count)
        
        if np.isnan(cor):
            continue
        else : 
            result.append((store_name, '{:.2f}'.format(cor), temp_type))
            
    result.sort(key=lambda r: r[1], reverse=True)
    
    return result[:n]
```

```python
recommend_result = recommend('힐링슈퍼', matrix, 10, similar_type=True)

pd.DataFrame(recommend_result, columns=['상호명', '상관도', '업종명'])
```
