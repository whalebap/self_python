```python
import numpy as np
import pandas as pd
import json
```

```python
# load dataset

meta = pd.read_csv('data/movies_metadata.csv')
meta.head()
# 처음 다섯줄을 로드해서 데이터셋을 확인함 

# 장르, 무비아이디(사용자 평가 데이터와 연결), 제목 등이 필요함
```

```python
# 필요한 칼럼만 놔두자

meta=meta[['movieid', 'original_title', 'original_language', 'genres']]
meta=meta.rename(columns={'movieid':'movieId'}) #칼럼명 바꾸기
meta=meta[meta['original_language']=='en'] #영어로 된 영화만 분석한다
meta.head()
```

```python
# 사용자의 평가 데이터 가져오기. 기존 데이터가 커서 ratings_small이란 
# 축약판 가져온다.

ratings = pd.read_csv('data/ratings_small.csv')
ratings = ratings[['userId', 'movieId', 'rating']]
ratings.head()
```

```python
ratings.describe()
# rating 테이블에 대한 정보 
# mean은 평균이니 사람들은 대충 3.5를 주는 군
```

```python
# string을 숫자로 바꿔주는 to_numeric 함수를 쓴다. 
meta.movieId = pd.to_numeric(meta.movieId, errors='coerce')
ratings.movieId=pd.to_numeric(ratings.movieId, errors='coerce')
```

```python
# 장르가 제이슨 형태이므로 계산하기 쉽게 배열 형태로 가공을 한ㄷ.

def parse_genres(genres_str):
    genres = json.loads(genres_str.replace('\'', '"'))
    # json 중 ''을 ""으로 교체
    
    genres_list=[]
    for g in genres:
        genres_list.append(g['name'])
        
    return genres_list 
#장르만 남겨서 장르 리스트를 만들었다. 

meta['genres'] = meta['genres'].apply(parse_genres)
# df.apply(def)는 각 행에 해당 함수를 적용한다. 
# 고로 각 행마다 parse_genres가 적용됨.
```

```python
# json에서 배열화됐고 이름만 담긴 장르들 확인. 
meta.head()
```

```python
# 두 개의 데이터프레임을 병합한다. 영화와 평가 테이블을 merge.
data=pd.merge(ratings, meta, on='movieId', how='inner')
# movieId를 기준으로 innerJoin 
data.head()

# 로키3 영화에 대한 모든 유저의 평가를 합침. 
# 사용자의 평가수x영화의 수 만큼 인스턴스가 생김.
```

```python
# 피벗 테이블을 만든다.

matrix=data.pivot_table(index='userId', columns='original_title', values='rating')
matrix.head(20) #원하는 튜플수만큼 

#  인덱스 유저 아이디, 칼럼 영화명, 값은 별점 
# 평가를 안 했으면 NaN이 뜸. 
```

```python
# 상관관계 구하기. 우리는 피어슨 상관관계를 쓴다. 

# 피벗테이블 매트릭스를 기준으로 나는 영화 B를 재밌게 봤다. 비슷한 영화를 보고 싶다.
# 내가 평가한 영화 1의 RATING 값을 S1, 영화2의 레이팅 값을 S2라 하자. 
# S1.MEAN(), S2.MEAN()은 S1과 S2의 전체 유저 RATING 평균값이다.
# 그 뒤에 영화1, 2의 각자 사람들 평점에서 평균값을 다 뺀다.=S1_C, S2_C
# 사람마다 있는 s1_c * s2_c 를 곱한다. 그게 영화간 상관관계다. 
# 양의 상관관계면 영화가 비슷, 음의 상관관계(둘 사이의 별점차이)가 크면 다른 영화
# 그 값을 모두 더 한다. 
# 양과 음의 상관관계를 선형적으로 표현한게 피어슨 상관관계다. 

GENRE_WEIGHT=0.1
def pearsonR(s1, s2):
    s1_c=s1-s1.mean()
    s2_c=s2-s2.mean()
    return np.sum(s1_c*s2_c) / np.sqrt(np.sum(s1_c **2) * np.sum(s2_c **2))
# 분자(np.sum(s1_c*s2_C)가 클수록 상관관계가 높게 나온다.

#  recommend 함수는 input으로 영화의 타이틀을 받고, 피벗테이블, 
# 몇개의 영화 추천받을지, 비슷한 장르에 가중치를 줄건지를 받음
def recommend(input_movie, matrix, n, similar_genre=True):
    input_genres=meta[meta['original_title'] == input_movie]['genres'].iloc(0)[0]
#     영화의 장르를 input_genres에 저장해줌

    result=[]
    for title in matrix.columns:
        if title==input_movie:
            continue
#    영화의 타이틀 순서대로 돌면서 
# 타이틀과 input_movie이 같단 건 똑같은 영화 추천이니 이건 제외시킴         
      
        cor = pearsonR(matrix[input_movie], matrix[title])
# 피어슨 상관관계를 계산한 결과가 cor
        
        if similar_genre and len(input_genres) >0:
#             장르의 길이가 0보다 클 때 장르끼리 비교한다. 
            temp_genres=meta[meta['original_title'] == title]['genres'].iloc(0)[0]
            
            same_count=np.sum(np.isin(input_genres, temp_genres))
            cor += (GENRE_WEIGHT * same_count)
# 지금 받아온 영화의 장르를 우리가 인풋으로 넣은 영화의 장르와 비교해서
# same_count=np.sum(np.isin(input_genres, tem_genres))
# np.isin() : 배열을 비교하여 똑같은 요소가 있으면 True를 반환한다.
# 같은 장르가 많을 수록 수가 커짐. 
#             cor += (GENRE_WEIGHT * same_count)
# 같은 장르가 많을 수록 장르에 가중치를 주게 됨. 
        
        if np.isnan(cor):
            continue
        else : 
            result.append((title, '{:.2f}'.format(cor), temp_genres))
#           결과를 result에 담아준다. 

    result.sort(key=lambda r: r[1], reverse=True)
#     rating이 높은 순서대로 정렬 후 내림차순 순서대로 정렬함. 
    return result[:n] #추천받을 갯수인 n개 만큼만 돌려준다. 
```

```python
recommend_result = recommend('The Dark Knight', matrix, 10, similar_genre=True)
# 다크나이트랑 비슷한 영화 10개 추천해. 장르가 비슷한 것도 좋아. 

pd.DataFrame(recommend_result, columns = ['Title', 'Correlation', 'Genre'])
# pandas에 데이트프레임으로 만들어서 정렬시킴. 
```
